```python
class Clinic:
    def __init__(self):
        self.data = []

    def book(self):
        name = input("Name: ")
        mobile = input("Mobile: ")
        doctor = input("Doctor: ")
        time = input("Time (10/11/12/2/3): ")

        count = 0
        for i in self.data:
            if i[2] == doctor and i[3] == time:
                count += 1

        if count < 3:
            self.data.append([name, mobile, doctor, time])
            print("Booked")
        else:
            print("Slot Full")

    def view_cancel(self):
        mobile = input("Mobile: ")
        for i in self.data:
            if i[1] == mobile:
                print(i)
                self.data.remove(i)
                print("Cancelled")
                return
        print("Not Found")

c = Clinic()
while True:
    print("1 Book 2 View/Cancel 3 Exit")
    ch = input()
    if ch == "1":
        c.book()
    elif ch == "2":
        c.view_cancel()
    else:
        break
```
