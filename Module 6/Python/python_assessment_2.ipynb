```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/Priya-tops/Tops-Assessment-Data/main/Data%20Analytics/Retail%20Data.csv"
df = pd.read_csv(url)

df.head()
```

```python
#Que 5
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)

```

```python
#Que 6
(df['Ship Date'] < df['Order Date']).sum()
```

```python
#Que 7
money_columns = ['Cost Price','Retail Price','Sub Total',
                 'Discount $','Order Total','Shipping Cost','Total']

for col in money_columns:
    df[col] = df[col].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
#Que 14
df['Sub Total'] = pd.to_numeric(df['Sub Total'], errors='coerce')
df['Discount %'] = pd.to_numeric(df['Discount %'], errors='coerce')
df['Discount $'] = pd.to_numeric(df['Discount $'], errors='coerce')

((df['Sub Total'] * df['Discount %'] / 100) - df['Discount $']).abs().sum()
```

```python
#Que 15
df['Calculated Total'] = df['Sub Total'] - df['Discount $'] + df['Shipping Cost']

(df['Calculated Total'] - df['Total']).abs().sum()
```
